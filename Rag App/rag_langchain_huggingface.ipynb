### Initialize HuggingFace LLM

```python
!pip install --upgrade transformers>=4.41.0 langchain-huggingface torch accelerate numpy

import os
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain_huggingface import HuggingFacePipeline

HF_TOKEN = "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_blYRVQZQomNmESsygRgycVJqxgHyeuXhFz"
os.environ["HF_TOKEN"] = HF_TOKEN

model_id = "mistralai/Mistral-7B-v0.1"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    dtype=torch.float16,
    low_cpu_mem_usage=True
)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=256,
    temperature=0.1,
    do_sample=True,
    return_full_text=False
)

llm = HuggingFacePipeline(pipeline=pipe)

# Correct way to generate text
result = llm.pipeline("Explain RAG in simple terms.")  # <- use .pipeline
print(result)
```

### Initialize Embedding Model

Model url: https://sbert.net/

```python
from langchain.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
  model_name="sentence-transformers/all-mpnet-base-v2"
)
```

### Initialize Output Parser

```python
from langchain.schema.output_parser import StrOutputParser

output_parser=StrOutputParser()
```

### Load PDF Document

```python
!pip install pypdf -qU
```

```python
from langchain_community.document_loaders import PyPDFLoader

# Load the PDF document
loader = PyPDFLoader("/content/AgstatBK.pdf")

docs = loader.load()
```

```python
len(docs)
```

```python
docs[0]
```

### Split Documents into Chunks

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Initialize the text splitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=50)

# Split the documents into chunks
splits = text_splitter.split_documents(docs)
```

```python
len(splits)
```

### Create Vector Store and Retriever

```python
from langchain_chroma import Chroma

# Create a vector store from the document chunks
vectorstore = Chroma.from_documents(documents=splits, embedding=embedding_model)
```

```python
# Create a retriever from the vector store
retriever = vectorstore.as_retriever()
```

### Define Prompt Template

```python
from langchain.prompts import ChatPromptTemplate

# Define prompt template
template = """
Answer this question using the provided context only.

{question}

Context:
{context}

Answer:
"""

prompt=ChatPromptTemplate.from_template(template)
```

```python
prompt
```

### Chain Retriever and Prompt Template with LLM

```python
from langchain.schema.runnable import RunnablePassthrough

chain = (
    {"context": retriever,  "question": RunnablePassthrough()}
    | prompt
    | llm
    | output_parser
)
```

#### Invoke RAG Chain with Example Questions

```python
response = chain.invoke("who is codeprolk?")
print(response)
```

```python
response = chain.invoke("what are the courses they offer")
print(response)
```

```python
response = chain.invoke("what are the popular videos in codeprolk youtube channel")
print(response)
```
