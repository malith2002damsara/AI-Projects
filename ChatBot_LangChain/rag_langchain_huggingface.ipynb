```python
import os
import requests
from dotenv import load_dotenv

# 🌟 Load environment variables from .env
load_dotenv()

GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
if not GEMINI_API_KEY:
    raise ValueError("GEMINI_API_KEY not found in .env")
os.environ["GEMINI_API_KEY"] = GEMINI_API_KEY

# ✅ Function to generate text from Gemini API
def generate_text(prompt, model="gemini-flash-latest", max_tokens=256):
    url = f"https://generativeai.googleapis.com/v1beta2/models/{model}:generateText"
    
    headers = {
        "Authorization": f"Bearer {GEMINI_API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "prompt": {"text": prompt},      # Gemini expects {"text": "..."} format
        "maxOutputTokens": max_tokens
    }

    try:
        response = requests.post(url, json=payload, headers=headers, timeout=15)
        response.raise_for_status()  # Raise error if HTTP status != 200
    except requests.exceptions.RequestException as e:
        return f"Request failed: {e}"

    data = response.json()
    # Gemini text is under candidates[0].content
    return data.get("candidates", [{}])[0].get("content", "")

# 🔹 Test
prompt = "Explain RAG in simple terms."
result = generate_text(prompt)
print("Generated Text:\n", result)
```
