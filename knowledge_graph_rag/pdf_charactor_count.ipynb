## PDF Text Character Counter

```python
%pip install -q pypdf

from pathlib import Path
from pypdf import PdfReader

ROOT_PATH = Path(r"D:\MyProjects\AI Project\knowledge_graph_rag")
pdf_paths = sorted(ROOT_PATH.rglob("*.pdf"))

print(f"Root path: {ROOT_PATH}")
print(f"PDF files found: {len(pdf_paths)}")
for p in pdf_paths:
    print(f"- {p}")
```

```python
def count_pdf_text_characters(pdf_path: Path) -> int:
    reader = PdfReader(str(pdf_path))
    text_parts = []
    for page in reader.pages:
        page_text = page.extract_text() or ""
        text_parts.append(page_text)
    return len("".join(text_parts))

results = []
for pdf in pdf_paths:
    try:
        char_count = count_pdf_text_characters(pdf)
        results.append({"pdf": str(pdf), "char_count": char_count, "status": "ok"})
    except Exception as e:
        results.append({"pdf": str(pdf), "char_count": 0, "status": f"error: {e}"})

results
```

```python
total_chars = sum(r["char_count"] for r in results if r["status"] == "ok")
ok_count = sum(1 for r in results if r["status"] == "ok")
err_count = sum(1 for r in results if r["status"].startswith("error"))

print(f"Processed PDFs: {ok_count}")
print(f"Failed PDFs: {err_count}")
print(f"Total extracted text characters: {total_chars}")

print("\nPer-file counts:")
for row in sorted(results, key=lambda x: x["char_count"], reverse=True):
    print(f"{row['char_count']:>10} | {row['status']:<10} | {row['pdf']}")
```
